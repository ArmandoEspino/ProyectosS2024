```python




























































from scipy.optimize import minimize

def objective_function(x):
    r, h = x
    return (r**2 * h) + 2 * (r**2)  # Costo total de fabricación

def volume_constraint(x):
    r, h = x
    return (3.14 * (r**2) * h) - 100  # Volumen fijo de la lata (por ejemplo, 100)

def aspect_ratio_constraint(x):
    r, h = x
    return h - (2 * r)  # Relación de aspecto constante (por ejemplo, 2)

# Valores iniciales para r y h
initial_guess = [1, 2]

# Definir las restricciones
constraints = [{'type': 'ineq', 'fun': volume_constraint},
               {'type': 'eq', 'fun': aspect_ratio_constraint}]

# Resolver el problema de optimización
solution = minimize(objective_function, initial_guess, constraints=constraints)

# Mostrar resultados
print("Valores óptimos:")
print("Radio (r):", solution.x[0])
print("Altura (h):", solution.x[1])
print("Costo mínimo:", solution.fun)
```
