```python
import pandas as pd
import numpy as np

# Carga de datos desde Restaurant Scrapped
data = pd.read_csv('Restaurant Scrapped.csv')

# Convertir la columna 'Rating' a tipo numérico
data['Rating'] = pd.to_numeric(data['Rating'], errors='coerce')

# Resume de datos
summary = data.describe()
print(summary)

# Media y desviación estándar de las calificaciones
mean_rating = data['Rating'].mean()
std_dev_rating = data['Rating'].std()
print(f"Media de calificación: {mean_rating}")
print(f"Desviación estándar de calificación: {std_dev_rating}")

# Visualización de datos
import matplotlib.pyplot as plt

# Histograma de calificaciones
plt.hist(data['Rating'], bins=5, edgecolor='black')
plt.xlabel('Calificación')
plt.ylabel('Frecuencia')
plt.title('Distribución de Calificaciones')
plt.show()

# Diagrama de cajita de calificaciones
plt.boxplot(data['Rating'])
plt.ylabel('Calificación')
plt.title('Diagrama de Caja de Calificaciones')
plt.show()
```
